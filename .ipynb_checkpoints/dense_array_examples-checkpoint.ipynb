```python
"""
Created on Wed Feb  5 21:54:43 2025

@author: Hilary Chang (hilarych@mit.edu)
"""
from obspy.core import Stream
from obspy import read,read_events
from glob import glob
from obspy import UTCDateTime
import pandas as pd
from obspy.taup.taup_geo import calc_dist
from obspy.geodetics import degrees2kilometers

df = pd.read_csv('Catalogs/stations_lasso_master.csv') # Station info
```

## 1. A regional event

```python
#%% 1. A regional event
evdp=6.09 # Event depth
evla=35.74 # Event latitude
evlo=-97.18 # Event Longitude 
t0= UTCDateTime(2016,4,27,15,44,55) # Event origin time
# USGS: https://earthquake.usgs.gov/earthquakes/eventpage/us10005by1/executive

file_list=glob('Data/20160427154455_OkM3.7_3min/*.sac')

# Initialize
st=Stream()
for file in file_list:
    tr=read(file)[0]
    print(tr)
    
    stn=int(tr.stats.station) # Station name
    # Add header info
    stla=df.loc[df['Station']==stn]['Lat'].values[0]
    stlo=df.loc[df['Station']==stn]['Lon'].values[0]
    tr.stats.sac['stla']=stla
    tr.stats.sac['stlo']=stlo

    tr.stats.sac['evdp']= evdp
    tr.stats.sac['evla']= evla
    tr.stats.sac['evlo']= evlo
    tr.stats.sac['t0']= t0
    
    # Distance to the event
    distance_deg=calc_dist(evla, evlo, stla,stlo,
                               radius_of_planet_in_km=6371,flattening_of_planet=0)
    tr.stats['distance']=degrees2kilometers(
                            degrees=distance_deg)*1000. # in meters
    st.append(tr)
```

```python
#%%% Examine the wavefield
st[0].stats # Header info
st[0].stats.sac.stlo # Station location 
st[0].stats.sac.stla

st[0].plot() # Plot recording at a station
st[1].plot() # Plot another station

st[0].data # Access the data
st[0].stats.sampling_rate # Sampling rate

st[::50].plot(type='section') # Plot the gather
```

## 2. A local event

```python
#%% 2. A local event
# Catalog for a local event
cat = read_events("Catalogs/20160416184918.xml")

cat[0]
cat[0].origins[0]
t0=cat[0].origins[0].time
evla=cat[0].origins[0].latitude
evlo=cat[0].origins[0].longitude
evdp=cat[0].origins[0].depth

file_list=glob('Data/p1_20160416184918_Main_2min/*.sac')

# Initialize
st=Stream()
for file in file_list:
    tr=read(file)[0]
    print(tr)
    
    stn=int(tr.stats.station) # Station name
    # Add header info
    stla=df.loc[df['Station']==stn]['Lat'].values[0]
    stlo=df.loc[df['Station']==stn]['Lon'].values[0]
    tr.stats.sac['stla']=stla
    tr.stats.sac['stlo']=stlo

    tr.stats.sac['evdp']= evdp
    tr.stats.sac['evla']= evla
    tr.stats.sac['evlo']= evlo
    tr.stats.sac['t0']= t0
    
    # Distance to the event
    distance_deg=calc_dist(evla, evlo, stla,stlo,
                               radius_of_planet_in_km=6371,flattening_of_planet=0)
    tr.stats['distance']=degrees2kilometers(
                            degrees=distance_deg)*1000. # in meters
    st.append(tr)
```
